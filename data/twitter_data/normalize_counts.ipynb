```python
import pandas as pd

# CSV einlesen
final_daily_df = pd.read_csv("final_daily_df.csv", parse_dates=["date"])

# Liste der zu normalisierenden Spalten
cols_to_normalize = ["tweet_count", "nlp_tweet_count"]

# Min-Max-Normalisierung (Überschreibt Originalwerte)
for col in cols_to_normalize:
    min_val = final_daily_df[col].min()
    max_val = final_daily_df[col].max()
    final_daily_df[col] = (final_daily_df[col] - min_val) / (max_val - min_val)

# CSV überschreiben
final_daily_df.to_csv("final_daily_df.csv", index=False)

# Kontrollausgabe
print(final_daily_df[cols_to_normalize].head())
```

```python
import pandas as pd
import matplotlib.pyplot as plt


# Plot tweet_count over time
plt.figure(figsize=(12, 6))
plt.plot(final_daily_df["date"], final_daily_df["tweet_count"], label="Tweet Count", color="blue")
plt.title("Tweet Count Over Time")
plt.xlabel("Date")
plt.ylabel("Tweet Count")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```
